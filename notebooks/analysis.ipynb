# Analysis — Article 4 §10

Loads `runs/*/summary.json` and produces the headline plot: violations + ambiguity by architecture, with information-loss correlation for MULTI-PIPE.

```python
import json
from pathlib import Path
import pandas as pd

rows = []
for summary in Path('../runs').glob('*/summary.json'):
    for entry in json.loads(summary.read_text()):
        rows.append({**entry, 'run_root': summary.parent.name})
df = pd.DataFrame(rows)
df
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
df.groupby('arch')['violations'].mean().plot(kind='bar', ax=ax[0], title='Mean violations')
df.groupby('arch')['ambiguity'].mean().plot(kind='bar', ax=ax[1], title='Mean residual ambiguity')
for a in ax:
    a.set_xlabel('')
fig.tight_layout()
```
